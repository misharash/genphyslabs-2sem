```python
from numpy import *
from scipy import genfromtxt
%matplotlib inline
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
data=genfromtxt("exp.csv",delimiter=",")
```

```python
U0=data[1:,0]
t=data[0,1:]
Un=data[1:,1:]
sUn=0.001
len(Un)
```

```python
Q=zeros((len(Un),len(Un)))
sQ=zeros((len(Un),len(Un)))
Rn=zeros((len(Un),len(Un)))
sRn=zeros((len(Un),len(Un)))
```

```python
for i in range(len(Un)):
    for j in range(len(Un[0,:])):
        Q[i,j]=.1*U0[i]*Un[i,j]
        sQ[i,j]=Q[i,j]*sUn/Un[i,j]
        Rn[i,j]=10*Un[i,j]/U0[i]
        sRn[i,j]=Rn[i,j]*sUn/Un[i,j]
```

```python
kQR=zeros(len(Un[0,:]))
R0=zeros(len(Un[0,:]))
skQR=zeros(len(Un[0,:]))
sR0=zeros(len(Un[0,:]))
def lin(x,a,b):
    return a*x+b
for i in range(len(Un[0,:])):
    plt.errorbar(Rn[:,i],Q[:,i],xerr=sRn[:,i],yerr=sQ[:,i],fmt='.')
    ko,sko=curve_fit(lin,Q[:,i],Rn[:,i],sigma=sRn[:,i])
    kQR[i],skQR[i],R0[i],sR0[i]=1/ko[0],sqrt(sko[0,0])/ko[0]**2,ko[1],sqrt(sko[1,1])
    plt.plot(Rn[:,i],(Rn[:,i]-R0[i])/ko[0])
    plt.xlabel("R, Ом")
    plt.ylabel("Q, Вт")
    plt.title('t={0:.0f}℃'.format(t[i]))
    plt.minorticks_on()
    plt.grid(which='both')
    plt.savefig("graph-QR{0:d}.pdf".format(i),format='pdf')
    plt.show()
print(kQR,skQR,R0,sR0,sep='\n')
```

```python
plt.errorbar(t,R0,yerr=sR0,fmt='.')
ko,sko=curve_fit(lin,t,R0,sigma=sR0,absolute_sigma=True)
R0C,sR0C,kRT,skRT=ko[1],sqrt(sko[1,1]),ko[0],sqrt(sko[0,0])
plt.plot(t,t*kRT+R0C)
plt.xlabel("t, ℃")
plt.ylabel("R, Ом")
plt.minorticks_on()
plt.axis([28,62,151,157])
plt.grid(which='both')
plt.savefig("graph-Rt.pdf",format='pdf')
alpha=kRT/R0C
salpha=alpha*sqrt((skRT/kRT)**2+(sR0C/R0C)**2)
print(alpha,salpha)
print(kRT,skRT)
print(R0C,sR0C)
```

```python
r1=0.05e-3
r2=10e-3
L=.367
kappa=kQR*kRT/(2*pi*L)*log(r2/r1)
skappa=kappa*sqrt((skQR/kQR)**2+(skRT/kRT)**2)
print(kappa,skappa,sep='\n')
```

```python
plt.errorbar(log(t+273),log(kappa),yerr=skappa/kappa,fmt=".")
ko,sko=curve_fit(lin,log(t+273),log(kappa),sigma=skappa/kappa,absolute_sigma=True)
plt.plot(log(t+273),log(t+273)*ko[0]+ko[1])
A,B,sA,sB=exp(ko[1]),ko[0],exp(ko[1])*sqrt(sko[1,1]),sqrt(sko[0,0])
plt.xlabel("ln T")
plt.ylabel("ln ϰ")
plt.minorticks_on()
plt.grid(which='both')
plt.savefig("graph-KT.pdf",format='pdf')
plt.show()
print(A,sA,B,sB)
```
