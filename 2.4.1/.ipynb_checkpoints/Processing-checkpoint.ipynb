```python
from numpy import *
from pylab import *
%matplotlib inline
```

```python
t,hd1,hd2=genfromtxt("exp.csv",delimiter=",",skip_header=1).T
len(t)
```

```python
h0=2.05
k=2*1.41/50
sh=1*k
h1=h0+k*hd1
h2=h0+k*hd2
h1
```

```python
oT=1/(273+t)
lP1=log(h1/h0)
slP1=sh/h1
lP2=log(h2/h0)
slP2=sh/h2
```

```python
from scipy.optimize import curve_fit
rcParams['figure.figsize']=(8,8)
def lin(x,a,b):
    return a*x+b
errorbar(oT*1e3,lP1,yerr=slP1,fmt=".",label="Нагревание")
errorbar(oT*1e3,lP2,yerr=slP2,fmt=".",label="Охлаждение")
ko,sko=curve_fit(lin,oT,lP1,sigma=slP1,absolute_sigma=False)
k1=ko[0]
sk1=sqrt(sko[0][0])
plot(oT*1e3,oT*ko[0]+ko[1],color="magenta",label="Нагревание")
print(k1,sk1)
ko,sko=curve_fit(lin,oT,lP2,sigma=slP2,absolute_sigma=False)
k2=ko[0]
sk2=sqrt(sko[0][0])
plot(oT*1e3,oT*ko[0]+ko[1],color="red",label="Охлаждение")
print(k2,sk2)
minorticks_on()
grid(which="both")
ylabel("ln(P/P₀)")
xlabel("1/T, 10⁻³ K⁻¹")
legend()
savefig("graph.pdf",format="pdf",dpi=300)
```
