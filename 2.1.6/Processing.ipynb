```python
from numpy import *
from scipy import optimize
from pylab import *
%matplotlib inline
```

```python
tt,Pd,U,sns=genfromtxt("exp.csv",delimiter=',',skip_header=1).T
```

```python
Pc=0.06
P=Pd*Pc
U0=array([-6.5]*7+[-6.5]*7+[-6.5]*7)
dT=(U-U0)/sns
#dT=U/sns
sP=Pc/2
Uc=1.
sU=Uc/2
sdT=sU/sns
```

```python
def f(x,a,b):
    return a*x+b
mu,smu,t=zeros((3,3))
for i in range(3):
    t[i]=tt[7*i]
    PP=P[7*i:7*(i+1)]
    sPP=sP
    dTT=dT[7*i:7*(i+1)]
    sdTT=sdT[7*i:7*(i+1)]
    errorbar(PP,dTT,xerr=sPP,yerr=sdTT,fmt='.')
    ko,sko=optimize.curve_fit(f,PP,dTT,sigma=sqrt(sPP**2+sdTT**2),absolute_sigma=False)
    mu[i],smu[i]=ko[0],sqrt(sko[0][0])
    print(mu[i],smu[i],ko[1],sqrt(sko[1][1]))
    PP=array(list(PP)+[0])
    plot(PP,PP*mu[i]+ko[1])
    minorticks_on()
    grid(which='both',color='grey',linestyle='-')
    xlabel("ΔP, атм")
    ylabel("ΔT, К")
    title("t={0:0.0f}℃".format(t[i]))
    savefig("graph{0:d}.pdf".format(i),dpi=300,format='pdf')
    show()
```

```python
R=8.31
CP=41
a,b,sa,sb=zeros((4,2))
for i in range(2):
    a[i]=(mu[i]-mu[i+1])/(1/(273+t[i])-1/(273+t[i+1]))*CP*R/2*1e-5
    sa[i]=a[i]*sqrt(smu[i]**2+smu[i+1]**2)/(mu[i]-mu[i+1])
    b[i]=CP*1e-5*(mu[i]*(273+t[i])-mu[i+1]*(273+t[i+1]))/(t[i+1]-t[i])
    sb[i]=b[i]*sqrt((smu[i]*(273+t[i]))**2+(smu[i+1]*(273+t[i+1]))**2)/(mu[i]*(273+t[i])-mu[i+1]*(273+t[i+1]))
    #b[i]=2*a[i]/R/(273+t[i])-mu[i+1]*CP*1e-5
    #sb[i]=2*sa[i]/R/(273+t[i])-smu[i+1]*CP*1e-5
```
